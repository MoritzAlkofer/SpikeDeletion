```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../Models/generalized_all_ref_loc/results_localized_Clemson.csv')
```

```python
locations = ['frontal', 'parietal', 'occipital', 'temporal', 'central','general']
```

```python
def plot_hist(ax,values,title):
    ax.hist(values,density=True)
    ax.set_xlim(0,1)
    ax.set_ylim(0,2)
    ax.set_title(title)
```

```python
def plot_hists(df,suptitle,locations):
    fig, axs = plt.subplots(2,3,figsize= (7,4),sharex=True,sharey=True)
    for i, location in enumerate(locations):
        preds = df[df.ChannelLocation==location].pred
        ax = axs[i//3,i%3]
        title = 'Channels: '+ location.capitalize()
        plot_hist(ax,preds,title)
    axs[0,0].set_ylabel('Prediction density')
    axs[1,0].set_ylabel('Prediction density')
    axs[1,0].set_xlabel('Prediction value')
    axs[1,1].set_xlabel('Prediction value')
    axs[1,2].set_xlabel('Prediction value')
    fig.suptitle(suptitle)
    plt.tight_layout()
```
